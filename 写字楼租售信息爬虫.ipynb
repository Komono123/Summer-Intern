```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def 查询():
    
    
    iteration = 1
    len_set = 0
    
    city = input("请输入需要查找的城市名称首字母缩写小写")
    key = input("请输入需要查找的城市地理位置：") #输入查询的地区名称
    url = f"https://{city}.office.fang.com/zu/house/i31-kw" + key
    
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    
    len_set = len(soup.find_all("div",{"class":"shop_no"}))
    
    if len_set > 0:
        
        print("该页面无搜索结果")
        
    else:
        #页面有返回搜寻结果
        while len_set == 0:


            url = f"https://{city}.office.fang.com/zu/house/i3{iteration}-kw" + key
            page = requests.get(url)
            soup = BeautifulSoup(page.content, "html.parser")

            price = soup.find_all("span",{"class":'red'})
            price_list = []
            for i in price:
                a = str(i.find("b")).split("<")[1].split(">")[1]
                price_list.append(a)


            size = soup.find_all("p", {"class":"tel_shop"})
            size_list = []
            for j in size:
                b = re.findall("[0-9]+", str(j))[1]
                size_list.append(b)


            name = soup.find_all("p",{"class":"add_shop"})
            name_list = []
            for i in name:
                index = str(i).find("title")
                a = str(i)[index:].split('"')[1]
                name_list.append(a)


            df = pd.DataFrame(list(zip(name_list,size_list, price_list)), columns = ['楼盘名称','平米数','每平米价格'])

            total_df.append(df)
            
            iteration += 1
            
            url1 = f"https://{city}.office.fang.com/zu/house/i3{iteration}-kw" + key
            page1 = requests.get(url1)
            soup1 = BeautifulSoup(page1.content, "html.parser")
            
            len_set = len(soup1.find_all("div",{"class":"shop_no"}))
    
    
    final = total_df[0]
    for i in range(1,len(total_df)):
        final = pd.concat([final, total_df[i]], axis = 0)
    final.reset_index(drop = True, inplace = True)
    
    return final
```
